```python
company = df['Company'].value_counts()

sns.set_style('darkgrid', {"grid.color": "0.6", "grid.linestyle": ":"})
company = company.sort_values(ascending=True)
company.plot(kind='barh',  title='Laptop manufacturer')
```

```python
Inches = df['Inches'].value_counts()

sns.set_style('darkgrid', {"grid.color": "0.6", "grid.linestyle": ":"})
Inches = Inches.sort_values(ascending=True)
Inches.plot(kind='barh', color='green', edgecolor='green', title='Screen Size')
```

```python
CPU_company = df['CPU_Company'].value_counts()

CPU_company.plot(kind='bar', color='black', edgecolor='black', width=0.3, title='CPU manufacturer')
```

```python
GPU_Company = df['GPU_Company'].value_counts()

GPU_Company.plot(kind='bar', color='blue', edgecolor='blue', width=0.3, title='GPU manufacturer')
plt.xticks(rotation=0)
plt.show()
```

```python
OpSys = df['OpSys'].value_counts()

plt.figure(figsize=(10,4))
OpSys.plot(kind='bar', color='blue', edgecolor='blue', title='Operating System')
plt.xticks(rotation=0)
plt.show()
```

```python
top_publishers = df.groupby('Company')['Price (Euro)'].sum().nlargest(8)
top_publishers.plot(kind='pie',    autopct='%.3f', title='Manufacturers selling price ratio')
plt.show()
```

```python
sns.barplot(x = df['OpSys'], y = df['Price (Euro)'])
plt.xticks(rotation='vertical')
plt.title('Operating System & Price (Euro)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Weight (kg)', y='Price (Euro)', data=df)
plt.title('Weight vs Laptop Price')
plt.xlabel('Weight (kg)')
plt.ylabel('Price (Euro)')
plt.show()
```

```python
x = df.corr(numeric_only=True)

sns.heatmap(x, annot=True, cmap='crest')

plt.show()
```

```python
company_avg_price = df.groupby('Company')['Price (Euro)'].mean()
plt.figure(figsize=(12, 6))
plt.bar(company_avg_price.index, company_avg_price.values, color='red')
plt.title('Average Laptop Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Price (Euro)')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_price_per_company = df.groupby('Company')['Price (Euro)'].mean()

plt.figure(figsize=(10, 6))
plt.bar(avg_price_per_company.index, avg_price_per_company.values)

plt.title('Average Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Price (Euro)')
plt.xticks(rotation=45)  
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for company in df['Company'].unique():
    subset = df[df['Company'] == company]
    plt.scatter(subset['Weight (kg)'], subset['Price (Euro)'], label=company)

plt.title('Price vs Weight by Company')
plt.xlabel('Weight (kg)')
plt.ylabel('Price (Euro)')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```
